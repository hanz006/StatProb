```python
import io
import pandas as pd
from google.colab import files
import matplotlib.pyplot as plt
import numpy as np
import math
from collections import Counter
from scipy.stats import kurtosis

print("Analisis Statistik Deskriptif Sederhana Menggunakan Python")
print("\n")

print("Anggota Kelompok : ")
print("1. Christiano Ronaldo Silalahi (5027241025)")
print("2. Mohamad Arkan Zahir Asyafiq (5027241120)")
print("3. Mohammad Abyan Ranuaji (5027241106)")
print("4. Aditya Reza Daffansyah (5027241034)")
print("5. S. Farhan Baig (5027241097)")
print("\n")


print("Tujuan:")
print("1. Mampu mengimpor dataset dalam format CSV dan XLSX menggunakan Python.")
print("2. Mampu menyajikan data menggunakan tabel distribusi frekuensi, histogram, dan poligon frekuensi.")
print("3. Mampu menerapkan konsep statistik deskriptif seperti ukuran pemusatan (mean, median, modus, kuartil)")
print("   dan ukuran persebaran (range, interquartile range, variansi, standar deviasi).")
print("4. Mampu melakukan analisis sederhana terkait outlier dan memberikan interpretasi hasil analisis.")
print("\n")

# Inisialisasi dataset
data = pd.read_excel('Dataset.xlsx')

# Data from the dataset (replace with your actual columns)
dataX = data['Curah Hujan (mm)'].tolist()
dataA = data['Penjualan A (pcs)'].tolist()
dataB = data['Penjualan B (pcs)'].tolist()


print("UKURAN PEMUSATAN DATA\n")

print("1. Mean")
# Mean untuk curah hujan
# Jumlah nilai dalam dataset
jumlah_nilaiX = sum(dataX)
print(f"Jumlah semua nilai curah hujan: {jumlah_nilaiX:.2f}")

# Jumlah total data dalam dataset
jumlah_total_dataX = len(dataX)
print(f"Jumlah total data curah hujan: {jumlah_total_dataX}")

# Jumlah semua nilai dalam dataset
mean_valueX = jumlah_nilaiX / jumlah_total_dataX
print(f"Mean curah hujan: {mean_valueX:.2f}\n")

#Mean untuk penjualan A
# Jumlah nilai dalam dataset
jumlah_nilaiA = sum(dataA)
print(f"Jumlah semua nilai penjualan A: {jumlah_nilaiA:.2f}")

# Jumlah total data dalam dataset
jumlah_total_dataA = len(dataA)
print(f"Jumlah total data penjualan A: {jumlah_total_dataA}")

# Jumlah semua nilai dalam dataset
mean_valueA = jumlah_nilaiA / jumlah_total_dataA
print(f"Mean penjualan A: {mean_valueA:.2f}\n")

#Mean untuk penjualan B
# Jumlah nilai dalam dataset
jumlah_nilaiB = sum(dataB)
print(f"Jumlah semua nilai penjualan B: {jumlah_nilaiB:.2f}")

# Jumlah total data dalam dataset
jumlah_total_dataB = len(dataB)
print(f"Jumlah total data penjualan B: {jumlah_total_dataB}")

# Jumlah semua nilai dalam dataset
mean_valueB = jumlah_nilaiB / jumlah_total_dataB
print(f"Mean penjualan B: {mean_valueB:.2f}\n\n")

print("2. Median")
# Mengurutkan data dari terkecil hingga terbesar
sorted_dataX = sorted(dataX)
print(f"Data berurut data curah hujan: {sorted_dataX}")

# Jumlah total data dalam dataset
n = len(sorted_dataX)
print(f"Jumlah total data curah hujan: {n}")

index_tengahX = n//2 # double slash // is used for floor division

# Jika jumlah data genap, median adalah rata-rata dari dua nilai tengah
if n % 2 == 0:
    print(f"Jumlah total data Genap")
    median1X = sorted_dataX[index_tengahX]
    median2X = sorted_dataX[index_tengahX - 1]
    print(f"Nilai tengah data curah hujan: {median1X} {median2X}")
    median_valueX = (median1X + median2X) / 2

# Jika jumlah data ganjil, median adalah nilai tengah
else:
    print(f"Jumlah total data Ganjil")
    median_valueX = sorted_dataX[index_tengahX]
    print(f"Nilai tengah data curah hujan: {median_valueX}")

print(f"Median Data curah Hujan : {median_valueX}\n")


# Mengurutkan data dari terkecil hingga terbesar
sorted_dataA = sorted(dataA)
print(f"Data berurut data penjualan barang A : {sorted_dataA}")

# Jumlah total data dalam dataset
n = len(sorted_dataA)
print(f"Jumlah total data penjualan barang A : {n}")

index_tengahA = n//2 # double slash // is used for floor division

# Jika jumlah data genap, median adalah rata-rata dari dua nilai tengah
if n % 2 == 0:
    print(f"Jumlah total data Genap")
    median1A = sorted_dataA[index_tengahA]
    median2A = sorted_dataA[index_tengahA - 1]
    print(f"Nilai tengah data penjualan barang A: {median1A} {median2A}")
    median_valueA = (median1A + median2A) / 2

# Jika jumlah data ganjil, median adalah nilai tengah
else:
    print(f"Jumlah total data Ganjil")
    median_valueA = sorted_dataA[index_tengahA]
    print(f"Nilai tengah data penjualan barang A: {median_valueA}")

print(f"Median data penjualan barang A : {median_valueA}\n")


# Mengurutkan data dari terkecil hingga terbesar
sorted_dataB = sorted(dataB)
print(f"Data berurut data penjualan barang B : {sorted_dataB}")

# Jumlah total data dalam dataset
n = len(sorted_dataB)
print(f"Jumlah total data penjualan barang B : {n}")

index_tengahB = n//2 # double slash // is used for floor division

# Jika jumlah data genap, median adalah rata-rata dari dua nilai tengah
if n % 2 == 0:
    print(f"Jumlah total data Genap")
    median1B = sorted_dataB[index_tengahB]
    median2B = sorted_dataB[index_tengahB - 1]
    print(f"Nilai tengah data penjualan barang B : {median1B} {median2B}")
    median_valueB = (median1B + median2B) / 2

# Jika jumlah data ganjil, median adalah nilai tengah
else:
    print(f"Jumlah total data Ganjil")
    median_valueB = sorted_dataB[index_tengahB]
    print(f"Nilai tengah data penjualan barang B : {median_valueB}")

print(f"Median data penjualan barang B: {median_valueB}\n\n")

print("3. Modus")
counterX = Counter(dataX)
mode_dataX = dict(counterX)  # ubah ke tipe data dictionary
max_counterX = max(list(counterX.values()))  # mencari data yang memiliki frekuensi kemunculan paling banyak

print(f"Max Counter: {max_counterX}")

mode_valueX = []
for k, v in mode_dataX.items():
    print(f"{k} muncul {v} kali")
    if v == max_counterX:  # cek jika frekuensi kemunculan sama dengan frekuensi maksimum
        mode_valueX.append(k)

if mode_valueX:
    print(f"Mode Data Curah Hujan : {mode_valueX}\n")
else:
    print("Tidak ada modus\n")

counterA = Counter(dataA)
mode_dataA = dict(counterA)  # ubah ke tipe data dictionary
max_counterA = max(list(counterA.values()))  # mencari data yang memiliki frekuensi kemunculan paling banyak

print(f"Max Counter: {max_counterA}")

mode_valueA = []
for k, v in mode_dataA.items():
    print(f"{k} muncul {v} kali")
    if v == max_counterA:  # cek jika frekuensi kemunculan sama dengan frekuensi maksimum
        mode_valueA.append(k)

if mode_valueA:
    print(f"Mode Data Penjualan Barang A: {mode_valueA}\n")
else:
    print("Tidak ada modus\n")


counterB = Counter(dataB)
mode_dataB = dict(counterB)  # ubah ke tipe data dictionary
max_counterB = max(list(counterB.values()))  # mencari data yang memiliki frekuensi kemunculan paling banyak

print(f"Max Counter: {max_counterB}")

mode_valueB = []
for k, v in mode_dataB.items():
    print(f"{k} muncul {v} kali")
    if v == max_counterB:  # cek jika frekuensi kemunculan sama dengan frekuensi maksimum
        mode_valueB.append(k)

if mode_valueB:
    print(f"Mode Data Penjualan Barang B: {mode_valueB}\n\n")
else:
    print("Tidak ada modus\n\n\n")

print("UKURAN PENYEBARAN DATA\n")
print("1. Range")
print("Range Data Curah Hujan")
# Fungsi untuk menghitung rentang
def calculate_range(dataX):
    nilai_tertinggi = max(dataX)
    print(f"Nilai tertinggi: {nilai_tertinggi}")
    nilai_terendah = min(dataX)
    print(f"Nilai terendah: {nilai_terendah}")
    # Selisih antara nilai tertinggi dan terendah
    data_range = max(dataX) - min(dataX)
    print(f"\n")
    print(f"Range: {nilai_tertinggi} - {nilai_terendah} = {data_range}")

    return data_range
data_range = calculate_range(dataX)
print("\n")


print("Range Data Penjualan Barang A")
def calculate_range(dataA):
    nilai_tertinggi = max(dataA)
    print(f"Nilai tertinggi: {nilai_tertinggi}")

    nilai_terendah = min(dataA)
    print(f"Nilai terendah: {nilai_terendah}")

    # Selisih antara nilai tertinggi dan terendah
    data_range = max(dataA) - min(dataA)

    print(f"\n")
    print(f"Range: {nilai_tertinggi} - {nilai_terendah} = {data_range}")

    return data_range
data_range = calculate_range(dataA)
print("\n")


print("Range Data Penjualan Barang B")
def calculate_range(dataB):
    nilai_tertinggi = max(dataB)
    print(f"Nilai tertinggi: {nilai_tertinggi}")

    nilai_terendah = min(dataB)
    print(f"Nilai terendah: {nilai_terendah}")

    # Selisih antara nilai tertinggi dan terendah
    data_range = max(dataB) - min(dataB)

    print(f"\n")
    print(f"Range: {nilai_tertinggi} - {nilai_terendah} = {data_range}")

    return data_range
data_range = calculate_range(dataB)
print("\n\n")


print("2. IQR (Interquartile Range)")
print("IQR Data Curah Hujan")
# Menghitung kuartil pertama (Q1) dan kuartil ketiga (Q3)
Q1 = np.percentile(dataX, 25)
Q3 = np.percentile(dataX, 75)

# Menghitung IQR
IQR = Q3 - Q1

# Output hasil
print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}\n")


print("IQR Data Penjualan Barang A")
# Menghitung kuartil pertama (Q1) dan kuartil ketiga (Q3)
Q1 = np.percentile(dataA, 25)
Q3 = np.percentile(dataA, 75)

# Menghitung IQR
IQR = Q3 - Q1

# Output hasil
print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}\n")

print("IQR Data Penjualan Barang B")
# Menghitung kuartil pertama (Q1) dan kuartil ketiga (Q3)
Q1 = np.percentile(dataB, 25)
Q3 = np.percentile(dataB, 75)

# Menghitung IQR
IQR = Q3 - Q1

# Output hasil
print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}\n\n")



print("3. Varians")
print("Varians Data Curah Hujan")
# Fungsi untuk menghitung rata-rata
def calculate_mean(dataX):
    return sum(dataX) / len(dataX)

# Fungsi untuk menghitung variansi
def calculate_variance(dataX):
    # Rata-rata
    mean = calculate_mean(dataX)
    print(f"Rata-rata: {mean}")

    # Kuadrat selisih antara setiap nilai data dengan rata-rata
    print(f"Kuadrat selisih setiap nilai data dengan rata-rata:")
    squared_diff = []
    for x in dataX:
        squared_diff_per_x = (x - mean)**2
        squared_diff.append( squared_diff_per_x )
        print(f"({x} - {mean})**2 = {(x - mean)}**2 = {squared_diff_per_x}")

    print(f"Total kuadrat selisih: {sum(squared_diff)}")
    print(f"Jumlah total data: {len(dataX)}")

    # Jumlah kuadrat selisih antara setiap nilai data dengan rata-rata, kemudian dibagi dengan jumlah total data
    variance = sum(squared_diff) / len(dataX)

    print(f"\n")
    print(f"Variansi: {sum(squared_diff)} / {len(dataX)} = {variance}")

    return variance
variance = calculate_variance(dataX)
print("\n")


print("Varians Data Penjualan Barang A")
# Fungsi untuk menghitung rata-rata
def calculate_mean(dataA):
    return sum(dataA) / len(dataA)

# Fungsi untuk menghitung variansi
def calculate_variance(dataA):
    # Rata-rata
    mean = calculate_mean(dataA)
    print(f"Rata-rata: {mean}")

    # Kuadrat selisih antara setiap nilai data dengan rata-rata
    print(f"Kuadrat selisih setiap nilai data dengan rata-rata:")
    squared_diff = []
    for x in dataA:
        squared_diff_per_x = (x - mean)**2
        squared_diff.append( squared_diff_per_x )
        print(f"({x} - {mean})**2 = {(x - mean)}**2 = {squared_diff_per_x}")

    print(f"Total kuadrat selisih: {sum(squared_diff)}")
    print(f"Jumlah total data: {len(dataA)}")

    # Jumlah kuadrat selisih antara setiap nilai data dengan rata-rata, kemudian dibagi dengan jumlah total data
    variance = sum(squared_diff) / len(dataA)

    print(f"\n")
    print(f"Variansi: {sum(squared_diff)} / {len(dataA)} = {variance}")

    return variance
variance = calculate_variance(dataA)
print("\n")


print("Varians Data Penjualan Barang B")
# Fungsi untuk menghitung rata-rata
def calculate_mean(dataB):
    return sum(dataB) / len(dataB)

# Fungsi untuk menghitung variansi
def calculate_variance(dataB):
    # Rata-rata
    mean = calculate_mean(dataB)
    print(f"Rata-rata: {mean}")

    # Kuadrat selisih antara setiap nilai data dengan rata-rata
    print(f"Kuadrat selisih setiap nilai data dengan rata-rata:")
    squared_diff = []
    for x in dataB:
        squared_diff_per_x = (x - mean)**2
        squared_diff.append( squared_diff_per_x )
        print(f"({x} - {mean})**2 = {(x - mean)}**2 = {squared_diff_per_x}")

    print(f"Total kuadrat selisih: {sum(squared_diff)}")
    print(f"Jumlah total data: {len(dataB)}")

    # Jumlah kuadrat selisih antara setiap nilai data dengan rata-rata, kemudian dibagi dengan jumlah total data
    variance = sum(squared_diff) / len(dataB)

    print(f"\n")
    print(f"Variansi: {sum(squared_diff)} / {len(dataB)} = {variance}")

    return variance
variance = calculate_variance(dataB)
print("\n\n")

print("4. Standar Deviasi")
# Fungsi untuk menghitung standar deviasi
def calculate_stddev(dataX):
    # Menghitung variansi terlebih dahulu
    varianceX = np.var(dataX)
    # Menghitung standar deviasi (akar kuadrat dari variansi)
    stddevX = np.sqrt(varianceX)
    return stddevX

# Menggunakan fungsi
stddevX = calculate_stddev(dataX)  # Menghitung standar deviasi dataX

# Output
print(f'Standar deviasi curah hujan : {stddevX:.2f}')



# Fungsi untuk menghitung standar deviasi
def calculate_stddev(dataA):
    # Menghitung variansi terlebih dahulu
    varianceA = np.var(dataA)
    # Menghitung standar deviasi (akar kuadrat dari variansi)
    stddevA = np.sqrt(varianceA)
    return stddevA

# Menggunakan fungsi
stddevA = calculate_stddev(dataA)  # Menghitung standar deviasi dataX

# Output
print(f'Standar deviasi penjualan A : {stddevA:.2f}')


# Fungsi untuk menghitung standar deviasi
def calculate_stddev(dataB):
    # Menghitung variansi terlebih dahulu
    varianceB = np.var(dataB)
    # Menghitung standar deviasi (akar kuadrat dari variansi)
    stddevB = np.sqrt(varianceB)
    return stddevB

# Menggunakan fungsi
stddevB = calculate_stddev(dataB)  # Menghitung standar deviasi dataX

# Output
print(f'Standar deviasi penjualan B: {stddevB:.2f}')
print("\n\n")



print("5. Skewness")
def skewness(mean_value, median_value, stddev):
    # Menghitung skewness
    skewnessX = 3 * (mean_value - median_value) / stddev
    return skewnessX  # Mengembalikan nilai skewness

# Menghitung skewness untuk dataX
skewnessX = skewness(mean_valueX, median_valueX, stddevX)
print("Nilai skewness untuk data curah hujan:", skewnessX)

# Menghitung skewness untuk dataA dan dataB
skewnessA = skewness(mean_valueA, median_valueA, stddevA)
skewnessB = skewness(mean_valueB, median_valueB, stddevB)

# Mencetak hasil skewness untuk dataA dan dataB
print("Nilai skewness untuk dataA:", skewnessA)
print("Nilai skewness untuk dataB:", skewnessB)
print("\n\n")

print("6. Kurtosis")
# Menghitung kurtosis untuk dataX
kurtosisX = kurtosis(dataX, axis=0, bias=True)
print(f"Kurtosis untuk Curah Hujan: {kurtosisX:.2f}")

# Menghitung kurtosis untuk dataA
kurtosisA = kurtosis(dataA, axis=0, bias=True)
print(f"Kurtosis untuk Penjualan A: {kurtosisA:.2f}")

# Menghitung kurtosis untuk dataB
kurtosisB = kurtosis(dataB, axis=0, bias=True)
print(f"Kurtosis untuk Penjualan B: {kurtosisB:.2f}")
print("\n\n\n")


print("VISUALISASI DATA\n")

print("1. Tabel Frekuensi")
# Fungsi untuk membuat tabel distribusi frekuensi
def buat_tabel_distribusi_frekuensi(data, nama_data):
    # Menghitung banyak kelas
    banyak_kelas = int(1 + 3.322 * np.log10(len(data))) # Pembulatan

    # Mencari nilai minimum dan maksimum
    nilai_min = min(data)
    nilai_max = max(data)

    # Menghitung interval kelas
    interval_kelas = round((nilai_max - nilai_min) / banyak_kelas)

    # Membuat bins untuk distribusi frekuensi
    bins = np.arange(nilai_min, nilai_max + interval_kelas, interval_kelas)

    # Membuat tabel distribusi frekuensi
    frekuensi, batas_bawah = np.histogram(data, bins=bins)

    # Membuat DataFrame untuk menampilkan tabel
    df = pd.DataFrame({
        'Batas Bawah': batas_bawah[:-1],
        'Batas Atas': batas_bawah[1:],
        'Frekuensi': frekuensi
    })

    print(f"Tabel Distribusi Frekuensi untuk {nama_data}:\n")
    print(df)
    print("\n")

# Membuat tabel distribusi frekuensi untuk setiap data
buat_tabel_distribusi_frekuensi(dataX, "Data Curah Hujan")
buat_tabel_distribusi_frekuensi(dataA, "Data Penjualan Barang A")
buat_tabel_distribusi_frekuensi(dataB, "Data Penjualan Barang B")
print("\n")


print("2. Histogram")
def buat_histogram(data, nama_data):
    plt.figure(figsize=(8, 5))  # Mengatur ukuran plot
    plt.hist(data, bins=10, color='orange', edgecolor='black')  # Membuat histogram
    plt.title(f'Histogram {nama_data}')  # Judul grafik
    plt.xlabel('Nilai')  # Label sumbu x
    plt.ylabel('Frekuensi')  # Label sumbu y
    plt.show()

# Membuat histogram untuk setiap data
buat_histogram(dataX, "Data Curah Hujan")
buat_histogram(dataA, "Data Penjualan Barang A")
buat_histogram(dataB, "Data Penjualan Barang B")
print("\n")

print("3. Poligon Frekuensi")
def buat_poligon_frekuensi(data, nama_data):
    # Menghitung frekuensi dan titik tengah
    frekuensi, bins = np.histogram(data, bins=10)
    bin_tengah = (bins[:-1] + bins[1:]) / 2

    # Membuat poligon frekuensi
    plt.figure(figsize=(8, 5))
    plt.plot(bin_tengah, frekuensi, marker='o', linestyle='-', color='green', label=f'{nama_data}')
    plt.title(f'Poligon Frekuensi {nama_data}')
    plt.xlabel('Nilai')
    plt.ylabel('Frekuensi')
    plt.legend()
    plt.show()

# Membuat poligon frekuensi untuk setiap data
buat_poligon_frekuensi(dataX, "Data Curah Hujan")
buat_poligon_frekuensi(dataA, "Data Penjualan Barang A")
buat_poligon_frekuensi(dataB, "Data Penjualan Barang B")
print("\n\n")

print("\nPENJELASAN BOXPLOT DALAM PYTHON\n")
print("Boxplot adalah representasi grafis dari suatu dataset yang menampilkan distribusi nilai-nilai dalam dataset tersebut.")
print("Ini berguna untuk memvisualisasikan dan membandingkan distribusi beberapa dataset.\n")



data = [23, 11, 18, 25, 9, 26, 17, 20, 16, 2]

plt.boxplot(data)

plt.show()

print("\nKotak mewakili rentang kuartil (IQR), yang merupakan perbedaan antara kuartil ke-75 (Q3) dan kuartil ke-25 (Q1).\nMedian diwakili oleh garis di dalam kotak.\nWhiskers memanjang dari kotak untuk menunjukkan rentang data.\nOutlier diwakili oleh titik-titik individu di luar kumis.\n")

print("PERBEDAAN BOXPLOT DENGAN HISTOGRAM\n")

data = np.random.randn(100)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(data, bins=20, edgecolor='black')
plt.title('Histogram')

plt.subplot(1, 2, 2)
plt.boxplot(data)
plt.title('Boxplot')

plt.show()

print("\nHistogram menampilkan distribusi data secara grafis dengan menggunakan bar-bar yang mewakili frekuensi nilai-nilai dalam dataset.")
print("Boxplot menampilkan distribusi data secara grafis dengan menggunakan kotak dan kumis yang mewakili rentang kuartil dan outlier.\n")

print("Kelebihan histogram:")
print("1. Menampilkan distribusi data secara grafis.")
print("2. Dapat menampilkan pola data yang kompleks.\n")

print("Kelebihan boxplot:")
print("1. Menampilkan rentang kuartil dan outlier secara jelas.")
print("2. Dapat membandingkan distribusi data antara beberapa kelompok.\n")

print("Kekurangan histogram:")
print("1. Tidak dapat menampilkan outlier secara jelas.")
print("2. Dapat dipengaruhi oleh ukuran bar.\n")

print("Kekurangan boxplot:")
print("1. Tidak dapat menampilkan pola data yang kompleks.")
print("2. Dapat dipengaruhi oleh ukuran dataset.\n")

print("Analisis tentang outlier dan interpretasi hasil analisis\n")

print("Box Plot adalah sebuah metode untuk menampilkan ringkasan dari sebuah set data yang memiliki properti seperti minimum, kuartil pertama, median, kuartil ketiga, dan maksimum.")
print("Dalam box plot, sebuah kotak dibuat dari kuartil pertama ke kuartil ketiga, dan sebuah garis vertikal yang melalui kotak di median.")
print("Box plot dapat digunakan untuk mendeteksi outlier dalam sebuah data.")
print("Outlier adalah sebuah nilai yang jauh berbeda dari nilai lainnya dalam sebuah data.")
print("Dalam analisis, box plot dapat membantu kita untuk mengidentifikasi outlier dan membuat interpretasi hasil analisis.")
print("Contoh penggunaan box plot untuk analisis outlier adalah sebagai berikut:\n")

np.random.seed(10)
data = np.random.normal(100, 20, 200)

fig = plt.figure(figsize =(10, 7))

plt.boxplot(data)

plt.show()

print("\nPlot di atas menampilkan box plot dari data yang dihasilkan oleh fungsi numpy.random.normal().")
print("Box plot ini terdiri dari kotak yang menampilkan kuartil pertama, median, dan kuartil ketiga, serta garis vertikal yang menunjukkan rentang data.")
print("Dalam plot ini, kita dapat melihat bahwa terdapat beberapa outlier yang jauh berbeda dari nilai lainnya dalam data.")
print("Plot box membuat lebih mudah untuk mengidentifikasi outlier dan membuat interpretasi hasil analisis.")


```
